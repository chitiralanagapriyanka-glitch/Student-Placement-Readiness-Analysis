```python
from plotly.subplots import make_subplots
import plotly.graph_objects as go

fig = make_subplots(
    rows=2, cols=2,
    subplot_titles=("CGPA Distribution",
                    "Coding Practice Hours",
                    "Internship Status",
                    "Certifications"),
    specs=[[{}, {}],
           [{'type':'domain'}, {}]]
)

# Histogram for CGPA
fig.add_trace(
    go.Histogram(
        x=df['2.  Current CGPA']
    ),
    row=1,col=1
)

# Bar chart for coding hours
coding=df["4. How many hours do you practice coding per day?"].value_counts()

fig.add_trace(
    go.Bar(
        x=coding.index,
        y=coding.values
    ),
    row=1,col=2
)

# Pie chart for internships
intern=df["6. Have you completed any internships?"].value_counts()

fig.add_trace(
    go.Pie(
        labels=intern.index,
        values=intern.values
    ),
    row=2,col=1
)


# Bar chart for certifications
cert=df["7. How many certifications have you completed?"].value_counts()

fig.add_trace(
    go.Bar(
        x=cert.index,
        y=cert.values
    ),
    row=2,col=2
)


fig.update_layout(
    height=700,
    title="Student Skills Insights Dashboard"
)

fig.show()
```

**INSIGHTS**

**Academic Performance Analysis**

*   Most students maintain a good CGPA range, showing strong academic performance among the respondents.
*   Higher CGPA students may have better readiness for placement opportunities.


**Programming Language Knowledge**

*   Many students know at least 1-3 programming languages.
*  Improving programming skills can increase technical confidence.


**Coding Practice Analysis**

*   A large number of students spend less time practicing coding regularly.
*   Consistent coding practice is required to improve problem-solving skills.


**Project Experience**

*   Students with more academic or personal projects have better practical exposure.
*   Building real-world projects improves placement preparation.


**Internship Analysis**

*   Some students have completed internships, while many are still lacking industry experience.
* More internship participation can improve professional skills.  


**Certification Analysis**

*   Most students completed a few certifications.
*   Certifications help students learn new technologies and strengthen resumes.



**Communication Skills Analysis**

*   Students with higher communication ratings are more confident about placement interviews.
*   Soft skills improvement is important along with technical knowledge.



**Aptitude Practice Analysis**


*   Regular aptitude practice is followed by fewer students.
*   Increasing aptitude preparation can improve placement test performance.



**Coding Platform Usage**

*   Students who regularly use platforms like HackerRank, LeetCode, or CodeChef have better coding practice habits.  


**Placement Confidence Analysis**

*   Interview confidence depends on multiple factors like coding skills, projects, communication, and preparation.


**Overall Conclusion**

The analysis shows that students are improving their technical skills through programming, projects, and certifications. However, regular coding practice, internships, aptitude preparation, and interview skills need more focus to increase placement readiness.
